```python
houseData4 = pd.read_csv('house-prices-advanced-regression-techniques/train.csv')
houseData4.head()
```

```python
houseGrLivArea = pd.DataFrame(houseData4['GrLivArea'])
houseSalePrice = pd.DataFrame(houseData4['SalePrice'])

# Against Log Sale Price
houseLOGSalePrice = pd.DataFrame(np.log(houseData4['SalePrice']))
```

```python
sb.jointplot(x=houseGrLivArea, y=houseSalePrice, height=12, edgecolor='w').set_axis_labels("houseGrLivArea", "houseSalePrice")
sb.jointplot(x=houseGrLivArea, y=houseLOGSalePrice, height=12, edgecolor='w').set_axis_labels("houseGrLivArea", "houseLOG(SalePrice)");
```

```python
# Import LinearRegression model from Scikit-Learn
from sklearn.linear_model import LinearRegression

# Create a Lienar Regression object
linreg = LinearRegression()
linreg_RTTS = LinearRegression()
```

```python
houseGrLivArea.shape, houseSalePrice.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Split dataset into Train & Test
houseGrLivArea_train = pd.DataFrame(houseGrLivArea[:1100])
houseGrLivArea_test = pd.DataFrame(houseGrLivArea[-360:]) # [(1100-1460):] || [1100:]
houseSalePrice_train = pd.DataFrame(houseSalePrice[:1100])
houseSalePrice_test = pd.DataFrame(houseSalePrice[-360:])

# Random Train-Test Split using "houseLOGSalePrice" # Either int(number)/float(%tage)
x_train, x_test, y_train, y_test = train_test_split(houseGrLivArea, houseLOGSalePrice, test_size = 360) 


# Check sample sizes
print("Train Set :", houseGrLivArea_train.shape, houseSalePrice_train.shape)
print("Test Set :", houseGrLivArea_test.shape, houseSalePrice_test.shape, "\n")

print("('RTTS') LOG Train Set :", x_train.shape, y_train.shape)
print("('RTTS') LOG Test Set :", y_test.shape, y_test.shape)
```

```python
# Fit LinReg model on houseGrLivArea_train & houseSalePrice_train
linreg.fit(houseGrLivArea_train, houseSalePrice_train)

# Fit LinReg model on RTTS houseGrLivArea_train & RTTS houseLOGSalePrice_train
linreg_RTTS.fit(x_train, y_train)
```

```python
# Coeffs of LinReg model 
print('Intercept \t: b = ', linreg.intercept_)
print('Coefficients \t: a = ', linreg.coef_, "\n") # or linreg.coef_[0]

# Coeffs of RTTS LOG LinReg model 
print('Intercept \t: b = ', linreg_RTTS.intercept_)
print('Coefficients \t: a = ', linreg_RTTS.coef_) # or linreg.coef_[0]
```

```python
# Formula for LinReg line
regline_x = houseGrLivArea_train
regline_y = linreg.intercept_ + linreg.coef_ * houseGrLivArea_train

regline_RTTS_x = x_train
regline_RTTS_y = linreg_RTTS.intercept_ + linreg_RTTS.coef_ * x_train

# Plot LinReg line
f, axes = plt.subplots(1,2,figsize=(16,8))
axes[0].scatter(houseGrLivArea_train, houseSalePrice_train, edgecolors='white')
axes[0].plot(regline_x, regline_y, 'r:', linewidth=3)
axes[0].set_xlabel("SalePrice_train (0-1100)")
axes[0].set_ylabel("HouseGrLivArea_train (0-1100)")

axes[1].scatter(x_train, y_train, edgecolors='white')
axes[1].plot(regline_RTTS_x, regline_RTTS_y, 'r:', linewidth=3)
axes[1].set_xlabel("houseLOGSalePrice (RTTS)")
axes[1].set_ylabel("HouseGrLivArea_train (RTTS)")
plt.show()
```

```python
# Predict Response based on Predictor
# Predict 'SalePrice' values corresponding to 'GrLivArea' in the test dataset
houseSalePrice_test_pred = linreg.predict(houseGrLivArea_test)
houseLOGSalePrice_test_pred = linreg_RTTS.predict(x_test)

# Plot the Predictions 
f, axes = plt.subplots(1,2,figsize=(16,8))
axes[0].scatter(houseGrLivArea_test, houseSalePrice_test, color="g", edgecolors='w')
axes[0].scatter(houseGrLivArea_test, houseSalePrice_test_pred, color="red", s=10)
axes[0].set_xlabel("houseGrLivArea_test")
axes[0].set_ylabel("houseSalePrice_test_pred")

axes[1].scatter(x_test, y_test, color='g', edgecolors='w')
axes[1].scatter(x_test, houseLOGSalePrice_test_pred, color='red', s=10)
axes[1].set_xlabel("x_test")
axes[1].set_ylabel("houseLOGSalePrice_test_pred")
plt.show()
```

```python
# Goodness of Fit of the LinReg model => Check how good predictions are on "TRAIN SET"/"TEST SET"

# Metric: Explained Variance (R^2) on "TRAIN SET"
print("Exlained Variance (R^2) on 'TRAIN SET' \t:", 
      linreg.score(houseGrLivArea_train, houseSalePrice_train))
print("Exlained Variance (R^2) on 'RTTS_TRAIN SET' with log(SalePrice) \t:", 
      linreg_RTTS.score(x_train, y_train), "\n")

# Metric: Explained Variance (R^2) on "TEST SET"
print("Explained Variance (R^2) on 'TEST SET' \t:",
      linreg.score(houseGrLivArea_test, houseSalePrice_test))
print("Explained Variance (R^2) on 'RTTS_TEST SET' with log(SalePrice) \t:",
      linreg_RTTS.score(x_test, y_test))
```

```python
# Predict the Response corresponding to Predictors
y_train_pred = dectree.predict(x_trainn)

# Print Classification Accuracy 
print("Classification Accuracy \t:", dectree.score(x_trainn, y_trainn))

# Plot the 2-way Confusion Matirx
from sklearn.metrics import confusion_matrix
sb.heatmap(confusion_matrix(y_trainn, y_train_pred),
          annot=True, fmt=".0f", annot_kws={"size":18});
```

```python
from sklearn.metrics import confusion_matrix
train_confusion = confusion_matrix(y_trainn, y_train_pred)
# print(train_confusion)
TP, TN, FP, FN = train_confusion[1, 1], train_confusion[0, 0], train_confusion[0, 1], train_confusion[1, 0]

from sklearn import metrics
# Overall, how often is the classifier correct?
accuracy = (TP + TN) / float(TP + TN + FP + FN)
print("Classification Accuracy \t:", dectree.score(x_trainn, y_trainn)) # metrics.accuracy_score(y_trainn, y_train_pred)

# When a positive value is predicted, how often is the prediction correct?
precision = TP / float(TP + FP)
print("Classification Precision\t:", precision)

# When the actual value is positive, how often is the prediction correct?
sensitivity = TP / float(FN + TP)
print("Classification Sensitivity\t:", sensitivity)

# When the actual value is negative, how often is the prediction correct?
specificity = TN / (TN + FP)
print("Classification Specificity\t:", specificity)
```

Check how good the predictions are on the Test Set.    
<b>Metrics :</b> `Classification Accuracy` & `Confusion Matrix`.

```python
# Extract the two variables X_test and y_test
x_testt = pd.DataFrame(houseData_test['SalePrice'])
y_testt = pd.DataFrame(houseData_test['CentralAir'])

# Predict the Response corresponding to Predictors
y_test_pred = dectree.predict(x_testt)

# Print Classification Accuracy
print("Classification Accuracy \t:", dectree.score(x_testt, y_testt))

# Plot the 2-way Confusion Matrix 
from sklearn.metrics import confusion_matrix
sb.heatmap(confusion_matrix(y_testt, y_test_pred),
          annot=True, fmt=".0f", annot_kws={"size":18});
```

```python
from sklearn.metrics import confusion_matrix
test_confusion = confusion_matrix(y_testt, y_test_pred)
# print(test_confusion)
TP2, TN2, FP2, FN2 = test_confusion[1, 1], test_confusion[0, 0], test_confusion[0, 1], test_confusion[1, 0]

from sklearn import metrics
# Overall, how often is the classifier correct?
accuracy = (TP2 + TN2) / float(TP2 + TN2 + FP2 + FN2)
print("Classification Accuracy \t:", dectree.score(x_testt, y_testt)) # metrics.accuracy_score(y_trainn, y_train_pred)

# When a positive value is predicted, how often is the prediction correct?
precision = TP2 / float(TP2 + FP2)
print("Classification Precision\t:", precision)

# When the actual value is positive, how often is the prediction correct?
sensitivity = TP2 / float(FN2 + TP2)
print("Classification Sensitivity\t:", sensitivity)

# When the actual value is negative, how often is the prediction correct?
specificity = TN2 / (TN2 + FP2)
print("Classification Specificity\t:", specificity)
```

**Important** : Note the huge imbalance in the *False Positives* and *False Negatives* in the confusion matrix. 

>For Training Data, `False Positives=58` whereas `False Negatives=8`. 

>For Test Data, `False Positives=16` whereas `False Negatives=3`. 

This should not be surprising, this is a direct effect of the huge `Y` vs `N` imbalance in the `CentralAir` variable. 
> As `CentralAir = Y` was more likely in the data, *False Positives* are more likely too.
